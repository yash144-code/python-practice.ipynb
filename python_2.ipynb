### Question: Check if a string is a palindrome.

Write a Python program that asks the user to enter a string. Based on the input, determine if the string is a palindrome (reads the same forwards and backwards, ignoring case and spaces), and print an appropriate message.

```python
input_string = input("Enter a string: ")

# Clean the string: convert to lowercase and remove spaces
cleaned_string = input_string.replace(" ", "").lower()

# Check if the cleaned string is equal to its reverse
if cleaned_string == cleaned_string[::-1]:
    print(f'"{input_string}" is a palindrome.')
else:
    print(f'"{input_string}" is not a palindrome,')
```

✅ Problem 3: E-commerce offer  
Input purchase amount  
If amount ≥ 5000 → 20% discount  
Else if ≥ 2000 → 10% discount  
Else → no discount  
Print final payable amount.  



```python
purchase_amount= int(input("Enter Purchased amount:"))

if purchase_amount>=5000:
  print("20% dicount available")
  print(purchase_amount-(purchase_amount*20/100))
elif purchase_amount>= 2000:
  print("10% discount available")
  print(purchase_amount-(purchase_amount*10/100))
else:
  print("no discount")
```

```python
purchase_amount= int(input("Enter Purchased amount:"))

if purchase_amount >= 5000:
  discount_rate =0.20
elif purchase_amount >= 2000:
  discount_rate = 0.10
else:
  discount_rate = 0

discount = purchase_amount * discount_rate
final_amount = purchase_amount - discount

print("Discount applied:",discount)
print("Payable amount:",final_amount)
```
